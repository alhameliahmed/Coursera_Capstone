```python
import requests
```

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
soup = BeautifulSoup(requests.get(url).content, "html.parser")
```

# Scraping data from wikipedia

```python
tb = soup.find("tbody")
data = []
for row in tb.find_all("tr"):
    data.append(", ".join([td.text for td in row.find_all("td")]))
```

# Converting data to dataframe

```python
data = map(lambda s: s.strip(), data)
```

```python
df = pd.DataFrame([sub.split(",") for sub in data], columns=('PostalCode', 'Borough', 'Neighborhood'))
df = df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
df = df.reset_index(drop=True)
df
```

# Getting red of not assigned Borough

```python
df = df[~df.Borough.str.contains('Not assigned')] 
```

```python
df=df.reset_index(drop=True)
df
```

# grouping Neighborhood according to PostalCode

```python
df =df.groupby(['PostalCode', 'Borough'])['Neighborhood'].agg([('Neighborhood', ', '.join)]).reset_index()
df
```

# assigning Queen's Park to not assigned Neighborhood

```python
df.Neighborhood = df.Neighborhood.apply(lambda x: "Queen's Park" if 'Not assigned' in x else x)
```

# create dataframe of Borough that has toronto in them

```python
tor_data = df[df['Borough'].str.contains('Toronto')].reset_index(drop=True)
tor_data.head()
```

# displaying map of the neighborhood

```python
address = 'Toronto'

geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)

latitude = location.latitude
longitude = location.longitude
#latitude ='43.6532'
#longitude='-79.3832'
map_tor = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(tor_data['Latitude'], tor_data['Longitude'], tor_data['Neighborhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_tor)  
    
map_tor
```
